```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sm
from matplotlib.animation import FuncAnimation
import math

t = np.linspace(0, 10, 1001)
r = 2 + np.sin(6*t)
theta = 7*t + 1.2*np.cos(6*t)
x = [q * np.cos(w) for q in r for w in theta]
y = [q * np.sin(w) for q in r for w in theta]

fig1, ax1 = plt.subplots(subplot_kw={'projection': 'polar'})
ax1.set_rmax(1000)
ax1.plot(theta, r)

plt.show()

fig2 = plt.figure(figsize=[10,10])
ax2 = fig2.add_subplot(1,1,1)
ax2.plot(x,y)
plt.show()
```
